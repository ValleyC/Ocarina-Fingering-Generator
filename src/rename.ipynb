```python
import os

# Define the path to the directory containing the images
image_directory = '../images/G'

# Iterate over each file in the directory
for filename in os.listdir(image_directory):
    # Construct the full file path
    old_file_path = os.path.join(image_directory, filename)
    
    # Ensure the file is an image (optional, if you want to rename only certain file types)
    if filename.endswith('.png') or filename.endswith('.jpg') or filename.endswith('.jpeg'):
        # Create the new filename by adding 'G' in front of the original filename
        new_filename = 'G' + filename
        
        # Construct the full path for the new filename
        new_file_path = os.path.join(image_directory, new_filename)
        
        # Rename the file
        os.rename(old_file_path, new_file_path)
        
        # Print the rename operation (optional)
        print(f'Renamed: {old_file_path} -> {new_file_path}')
```

```python
import os

# Define the correct mapping from G key notes to F key notes
g_to_f_mapping = {
    'G-2': '-3',
    'G#-2': '-4',
    'G-3': '#-4',
    'G-4': '-5',
    'G#-4': '#-5',
    'G-5': '-6',
    'G#-5': '#-6',
    'G-6': '-7',
    'G#-6': '1',
    'G-7': '#1',
    'G1': '2',
    'G#1': '#2',
    'G2': '3',
    'G#2': '4',
    'G3': '#4',
    'G4': '5',
    'G#4': '#5',
    'G5': '6',
    'G#5': '#6',
    'G6': '7',
    'G#6': '+1'
}

# Define the path to the directory containing the images
image_directory = '../images/F'

# Iterate over each file in the directory
for filename in os.listdir(image_directory):
    # Construct the full file path
    old_file_path = os.path.join(image_directory, filename)
    
    # Extract the note name from the filename (assuming the filename is something like 'G1.png')
    note_name = filename.split('.')[0]  # Get the part before .png
    
    # Look up the corresponding F key note name
    if note_name in g_to_f_mapping:
        # Add 'F' before the mapped F key note
        new_note_name = 'F' + g_to_f_mapping[note_name]
        
        # Create the new filename
        new_filename = new_note_name + '.png'
        
        # Construct the full path for the new filename
        new_file_path = os.path.join(image_directory, new_filename)
        
        # Rename the file
        os.rename(old_file_path, new_file_path)
        
        # Print the rename operation (optional)
        print(f'Renamed: {old_file_path} -> {new_file_path}')
    else:
        print(f'No mapping found for {note_name}, skipping...')
```
